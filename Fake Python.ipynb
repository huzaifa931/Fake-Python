```python
import requests #to use http protocols
from bs4 import BeautifulSoup
```

```python
url = 'https://realpython.github.io/fake-jobs/'
request = requests.get(url)
request
```

```python
soup = BeautifulSoup(request.content,'html.parser')
soup
```

```python
title = soup.find('h2',class_ = 'title is-5')
print(title.text)
```

```python
sub_title = soup.find('h3',class_ = 'subtitle is-6 company')
print(sub_title.text)
```

```python
location = soup.find('p',class_ = 'location')
print(location.text)
```

```python
date = soup.find('p',class_ = 'is-small has-text-grey')
print(date.text)
```

```python
print(title.text.strip())  # removing space
print(sub_title.text.strip())
print(location.text.strip())
print(date.text.strip())
```

```python
divs = soup.find_all('div',class_='column is-half')
divs
```

```python
all_data = []

for div in divs:
    data = []
    title = soup.find('h2',class_ = 'title is-5').text.strip()
    sub_title = soup.find('h3',class_ = 'subtitle is-6 company').text.strip()
    location = soup.find('p',class_ = 'location').text.strip()
    date = soup.find('p',class_ = 'is-small has-text-grey').text.strip()
    data = [title,sub_title,location,date]
    all_data.append(data)

all_data
#save data in excel

    
```
